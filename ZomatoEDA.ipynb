```python
path="E:\ML\MachineLearningBasics\Datasets\Country-Code.xlsx"
df_country=pd.read_excel(path)
df.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
#To check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#Pie Chart(top 3 countries)
plt.pie(country_val[0:3],labels=country_names[0:3],autopct='%1.2f%%')
```

```python
#Observation
# Zomato maximum customer base in India
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
#Observartion
# when rating is between 4.5 to 4.9->excellent
# when rating is between 4.0 to 3.4->very good
# when rating is between 3.5 to 3.9->good
# when rating is between 3.0 to 3.4->average
# when rating is between 2.5 to 2.9->average
# when rating is between 2.0 to 2.4->Poor 
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['grey','red','orange','yellow','green','green'])
```

```python
# Observations
# Not Rated Count is very high
# Maximum number of rating are between 2.5 to 3.4
```

```python
#Count Plot unique no of values
sns.countplot(x="Rating color",data=ratings,hue="Rating color",palette=['grey','red','orange','yellow','green','green'])
```
